# 04 - LoRA vs QLoRA Comparison and Report

This notebook provides final side-by-side comparison and loads the generated markdown report.

## Load summary metrics and deltas

```python
from pathlib import Path
import json
import pandas as pd
import plotly.express as px

ROOT = Path.cwd().resolve()
if not (ROOT / 'pyproject.toml').exists():
    ROOT = ROOT.parent.resolve()
ART = ROOT / 'artifacts'

summary = json.loads((ART / 'metrics' / 'summary.json').read_text())
summary['deltas']
```

## Metrics table

```python
metrics = pd.read_csv(ART / 'metrics' / 'evaluation_metrics.csv')
metrics
```

## Accuracy chart

```python
fig = px.bar(metrics, x='run_name', y='accuracy', color='model_name', title='Accuracy by Run')
fig
```

## Macro F1 chart

```python
fig2 = px.bar(metrics, x='run_name', y='macro_f1', color='model_name', title='Macro F1 by Run')
fig2
```

## Generated markdown report (real output)

```python
report_path = ART / 'reports' / 'lora_qlora_report.md'
print(report_path)
print(report_path.read_text()[:2200])
```
